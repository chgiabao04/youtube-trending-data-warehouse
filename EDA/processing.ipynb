```python
import re
import html

def clean_text(text):
    text = str(text)
    text = html.unescape(text)           # Decode HTML
    text = text.lower()                  # Lowercase
    text = re.sub(r"http\S+|www\S+", "", text)  # Xóa URL
    text = re.sub(r"@\w+", "", text)     # Xóa mention
    text = re.sub(r"\d{9,}", "", text)   # Xóa số điện thoại
    text = re.sub(r"[^\w\s!?]", " ", text)  # Giữ chữ + ! + ?
    text = re.sub(r"\s+", " ", text).strip() # Xóa space thừa
    return text

df_comments["comment_text"] = df_comments["comment_text"].apply(clean_text)
df_comments = df_comments[df_comments["comment_text"].str.split().str.len() >= 2]

print("Shape sau clean:", df_comments.shape)
```
