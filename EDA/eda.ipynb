```python
corr = df_video[['views', 'likes', 'comment_count', 'duration', 'subscriber_count']].corr()

sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

- **views ↔ likes: 0.78 → rất cao, rất tốt cho mô hình**
- **likes ↔ comment_count: 0.42 → trung bình**
- views ↔ subscriber_count: 0.22 → thấp, channel lớn chưa chắc views cao
- views ↔ comment_count: 0.27 → thấp
- **duration ↔ views: -0.041 → tương quan yếu gần như không tương quan với views**
- **duration ↔ likes: -0.056 → tương tự**
- duration ↔ subscriber_count: 0.017 → gần như không liên quan

```python
import numpy as np

df_log = df_video.copy()
df_log['log_views']            = np.log1p(df_video['views'])
df_log['log_likes']            = np.log1p(df_video['likes'])
df_log['log_comment_count']    = np.log1p(df_video['comment_count'])
df_log['log_duration']         = np.log1p(df_video['duration'])
df_log['log_subscriber_count'] = np.log1p(df_video['subscriber_count'])

cols_corr = ['log_views', 'log_likes', 'log_comment_count', 
             'log_duration', 'log_subscriber_count']

corr = df_log[cols_corr].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn', center=0)
plt.title('Correlation Matrix (Log Scale)')
plt.tight_layout()
plt.show()
```

- log_likes            (0.81)  → feature mạnh nhất
- log_comment_count    (0.78)  → rất quan trọng
- log_subscriber_count (0.62)  → quan trọng
- log_duration         (-0.17) → yếu nhưng giữ, phân biệt Shorts/Long

```python
df_video["log_views"] = np.log1p(df_video["views"])
```

### Tạo Bucket Duration

```python
def bucket_duration(seconds):
    if seconds < 60:
        return "Short"
    elif seconds < 600:
        return "Medium"
    else:
        return "Long"

df_video['duration_bucket'] = df_video['duration'].apply(bucket_duration)
print(df_video["duration_bucket"].value_counts())
```

```python
sns.boxplot(x='duration_bucket', y='log_views', data=df_video)
```

## Views theo Category 

vì ban đầu k có crawl category_name nên sẽ tiến hành map lại với id

```python
category_map = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    15: "Pets & Animals",
    17: "Sports",
    19: "Travel & Events",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "Howto & Style",
    27: "Education",
    28: "Science & Technology"
}

df_video["category_name"] = df_video["category_id"].map(category_map)
print(df_video["category_name"].value_counts())
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_video, x='category_name', y=np.log1p(df_video['views']))
plt.title('Phân phối Views theo Category (Log Scale)')
plt.xlabel('Category')
plt.ylabel('log1p(views)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

có thể giữ lại Category_name kết hợp với topic_name để giúp ML học tốt hơn
