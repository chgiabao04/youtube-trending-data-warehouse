```python
import pandas as pd
```

```python
dim_video = pd.read_csv("warehouse/dim_video.csv")
dim_video = pd.DataFrame(dim_video)
display(dim_video)
```

```python
dim_video.info()
```

```python
dim_video.isnull().sum()
```

có nhiều video không để tags nên ta sẽ không xử lý nó

```python
dim_video = dim_video.dropna(subset=["duration"])
```

```python
import isodate

def parse_duration(x):
    try:
        return int(isodate.parse_duration(x).total_seconds())
    except:
        return None

dim_video["duration"] = dim_video["duration"].apply(parse_duration)
```

chuyển Duration thành giây

```python
dim_video["category_id"] = dim_video["category_id"].astype(int)
dim_video["keyword_id"] = dim_video["keyword_id"].astype(int)
dim_video["topic_id"] = dim_video["topic_id"].astype(int)
```

ép kiểu từ numeric thành int
